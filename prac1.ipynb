#### 7. [4 балла] Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года.

Предварительно создадим 2 ДатаФрейма с курсами групп 18-ого и 17-ого годов набора с помощью `pd.concat()` и удалим повторяющиеся строчки, чтобы получить уникальные названия курсов.

```python
# Создадим ДатаФреймы со всеми осенними и весенними курсами для групп
# 18-ого и 17-ого годов
courses_18 = pd.concat([df[df["18_group"] > 0]["fall_1"],
                        df[df["18_group"] > 0]["fall_2"],
                        df[df["18_group"] > 0]["fall_3"],
                        df[df["18_group"] > 0]["spring_1"],
                        df[df["18_group"] > 0]["spring_2"],
                        df[df["18_group"] > 0]["spring_3"]])\
    .rename("course")\
    .drop_duplicates()\
    .reset_index()\
    .drop(columns="index")

courses_17 = pd.concat([df[df["17_group"] > 0]["fall_1"],
                        df[df["17_group"] > 0]["fall_2"],
                        df[df["17_group"] > 0]["fall_3"],
                        df[df["17_group"] > 0]["spring_1"],
                        df[df["17_group"] > 0]["spring_2"],
                        df[df["17_group"] > 0]["spring_3"]])\
    .rename("course")\
    .drop_duplicates()\
    .reset_index()\
    .drop(columns="index")
```

Применим метод `merge()` с внутренним слиянием, чтобы найти пересечения, т.е. общие курсы для 17-ого и 18-ого года.

```python
# Внутреннее слияние
courses_18.merge(courses_17, how="inner", on="course")
```

Методом исключения найдите курсы, которые предлагались только студентам 18-го года и только студентам 17-го года.

Применим метод `merge()` с левосторонним слиянием и используем параметр метода `indicator=True`, чтобы после определить с помощью фильтра `left_only` строки, уникальные для 18-ого и 17-ого годов.

```python
# Левостороннее слияние для двух ДатаФреймов и использованиие параметра indicator=True
courses_18_unique = courses_18.merge(courses_17, how="left", on="course", indicator=True)\
    .rename(columns={"course": "course_18"})
courses_17_unique = courses_17.merge(courses_18, how="left", on="course", indicator=True)\
    .rename(columns={"course": "course_17"})
display(courses_18_unique[courses_18_unique["_merge"] == "left_only"]["course_18"]\
        .reset_index()\
        .drop(columns="index"),
    courses_17_unique[courses_17_unique["_merge"] == "left_only"]["course_17"]\
        .reset_index()\
        .drop(columns="index"))
```

### Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. Здесь будут описаны ключевые правила оформления графиков для **всех** домашних заданий.

У matplotlib, конечно же, есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторой величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить один из них. Не забывайте про базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах);
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать;
- если распределение на гистограмме имеет тяжёлые хвосты, лучше использовать логарифмическую шкалу.

Еще одна билиотека для визуализации: [seaborn](https://jakevdp.github.io/PythonDataScienceHandbook/04.14-visualization-with-seaborn.html) (обычно сокращается до sns). Это настройка над matplotlib, иногда удобнее и красивее делать визуализации через неё.

**5 пунктов после которых ваши графики не будут прежними:**
- для красивой картинки <code>%config InlineBackend.figure_format = 'retina'</code>
- задаем размер (почти) всех графиков (можно выбрать другие значения) <code>plt.rcParams['figure.figsize'] = 8, 5</code>
- размер шрифта подписей графиков <code>plt.rcParams['font.size'] = 12</code>
- формат в котором сохраняется изображение <code>mpl.rcParams['savefig.format'] = 'pdf'</code>

1. По умолчанию на каждой дисциплине по выбору у 3 и 4 курсов может учиться 1 группа (до 30 студентов). Исключения описаны ниже. На blended-дисциплинах для четверокурсников количество мест не ограничено.

Пройдемся циклом по уникальным значениям осенних дисциплин, создавая на каждой итерации отдельный датафрейм, представляющий из себя список поступающих студентов, отсортированный по перцентилю. Стоит отметить, что на данном этапе не производится зачисление студентов и создание таблицы студентов с курсами, на которые они зачислены (это будет выполнено на следующем шаге, делая итерацию по волнам).
```python
# Переберем уникальные значения дисциплин и создадим списки по ним
for i in df_fall_1["value"].unique():
    # Создаем списки по рейтингу каждой дисциплины
    df_fall_1_temp = df_fall_1.loc[df_fall_1["value"] == i]\
        .groupby(["value", "id"])["percentile"]\
        .mean()\
        .sort_values()\
        .reset_index()

    # Определяем длину списка для отбора
    stud_len = 0
    if i in ("Statistical Learning Theory", "Высокопроизводительные вычисления"):
        stud_len = 60
    elif i == "Анализ неструктурированных данных":
        stud_len = df_fall_1_temp.shape[0]
    else:
        stud_len = 30
    if stud_len > df_fall_1_temp.shape[0]:
        stud_len = df_fall_1_temp.shape[0]
    df_fall_1_temp = df_fall_1_temp.head(stud_len)

    # Помечаем первого и последнего студента в списке
    df.loc[df_fall_1_temp.iloc[0]["id"] == df["id"],
        "is_first_place"] = True
    df.loc[df_fall_1_temp.iloc[stud_len-1]["id"] == df["id"],
        "is_last_place"] = True
```

Создаем на каждой итерации таблицу с рейтингом, но в этот раз учитываем ещё и волну отбора. После создания списка зачисления перерасчитываем количество свободных мест для следующих волн. Самое важное, проверять, является ли студент уже полностью зачислен, если же да, удалим все строки с `id` этого студента из `df_fall`.

```python
# Создаем словарь с названиями дисциплин и количеством свободных мест на курсе
courses_available_places = {}
for index in df_fall["value"].unique():
    courses_available_places[index] = 30

# Меняем размер группы для определенных дисциплин
courses_available_places["Statistical Learning Theory"] = 60
courses_available_places["Высокопроизводительные вычисления"] = 60
# На этой дисциплине бесконечное количество мест,
# будем определять количество студентов в цикле
courses_available_places["Анализ неструктурированных данных"] = 0

# Итерация по волнам отбора
for i in range(1, 4):
    # Итерация по курсам
    for j in courses_available_places.keys():
        # Создаем временную таблицу с рейтингом по курсу
        df_fall_temp = df_fall.loc[(df_fall["value"] == j)
            & (df_fall["wave"] == i)]\
            .groupby(["value", "id"])["percentile"]\
            .mean()\
            .sort_values()\
            .reset_index()\

        # Определяем длину списка
        stud_len = courses_available_places[j]
        if stud_len > df_fall_temp.shape[0]:
            stud_len = df_fall_temp.shape[0]
        if j == "Анализ неструктурированных данных":
            stud_len = df_fall_temp.shape[0]

        # Меняем количество свободных мест в курсах после зачисления
        courses_available_places[j] = courses_available_places[j] - stud_len

        # Оставляем только необходимое количество человек
        df_fall_temp = df_fall_temp.head(stud_len)

        # Делаем слияние с ДатаФреймом осеннего зачисления
        fall_courses = fall_courses.merge(df_fall_temp[["id", "value"]], how="left")

        # Записываем зачисленных
        fall_courses.loc[(fall_courses["value"] == j)
            & (fall_courses["course1"] == "???"), "course1"] = j
        fall_courses.loc[(fall_courses["value"] == j)
            & (fall_courses["course1"] != "???")
            & (fall_courses["course1"] != j), "course2"] = j

        # Удаляем столбец value из fall_courses после слияния
        fall_courses = fall_courses.drop(columns="value")

        # Удаляем уже полностью зачисленных студентов
        df_fall = df_fall.merge(fall_courses.loc[(fall_courses["course1"] != "???")
            & (fall_courses["course2"] != "???")]["id"],
            how="left", on="id", indicator=True)
        df_fall = df_fall[df_fall["_merge"] == "left_only"].reset_index()\
            .drop(columns=["index", "_merge"])

# Сохраняем .csv файл
fall_courses.to_csv("res/res_fall.csv", index=False)
```

```python
# ДатаФрейм, в котором распределены волны отбора
df_spring = df[["id", "spring_1", "spring_2", "spring_3", "percentile",
                "spring_courses", "is_ml_student"]]
df_spring = df_spring.melt(id_vars=["id", "percentile",
                                    "spring_courses", "is_ml_student"],
                           value_vars=["spring_1", "spring_2", "spring_3"],
                           var_name = "spring")\
    .sort_values(by=["percentile", "id"])

# Учитываем смещение приоритетов
df_spring = df_spring[~((df_spring["is_ml_student"] == 1)
    & (df_spring["value"] == "Машинное обучение 2"))].copy()
df_spring.loc[:, "new_spring"] = df_spring.groupby("id").cumcount() + 1
df_spring = df_spring.drop(columns="spring")\
    .rename(columns={"new_spring": "spring"})

# Удалим повторы предметов в приоритетах
df_spring = df_spring.drop_duplicates(subset=["id", "value"])

# Добавим новый столбец wave, в который запишем в какой волне зачисления участвует студент
for i in range(1, 4):
    df_spring.loc[df_spring["spring"] == i, "wave"] = i
df_spring.loc[(df_spring["spring_courses"] == 2)
    & ((df_spring["spring"] == 1)
    | (df_spring["spring"] == 2)), "wave"] = 1
df_spring.loc[(df_spring["spring_courses"] == 2)
    & ((df_spring["spring"] == 3)), "wave"] = 2

# Удалим ненужные столбцы
df_spring = df_spring.drop(columns="spring_courses")

# Выводим полученный ДатаФрейм
df_spring
```

```python
# Создаем словарь с названиями дисциплин и количеством свободных мест на курсе
courses_available_places = {}
for index in df_spring["value"].unique():
    courses_available_places[index] = 30

# Меняем размер группы для определенных дисциплин
courses_available_places["Обучение с подкреплением"] = 60
courses_available_places["Анализ данных в бизнесе"] = 60

# Итерация по волнам отбора
for i in range(1, 4):
    # Итерация по курсам
    for j in courses_available_places.keys():
        # Создаем временную таблицу с рейтингом по курсу
        df_spring_temp = df_spring.loc[(df_spring["value"] == j)
            & (df_spring["wave"] == i)]\
            .groupby(["value", "id"])["percentile"]\
            .mean()\
            .sort_values()\
            .reset_index()\

        # Определяем длину списка
        stud_len = courses_available_places[j]
        if stud_len > df_spring_temp.shape[0]:
            stud_len = df_spring_temp.shape[0]

        # Меняем количество свободных мест в курсах после зачисления
        courses_available_places[j] = courses_available_places[j] - stud_len

        # Оставляем только необходимое количество человек
        df_spring_temp = df_spring_temp.head(stud_len)

        # Делаем слияние с ДатаФреймом осеннего зачисления
        spring_courses = spring_courses.merge(df_spring_temp[["id", "value"]], how="left")

        # Записываем зачисленных
        spring_courses.loc[(spring_courses["value"] == j)
            & (spring_courses["course1"] == "???"), "course1"] = j
        spring_courses.loc[(spring_courses["value"] == j)
            & (spring_courses["course1"] != "???")
            & (spring_courses["course1"] != j), "course2"] = j

        # Удаляем столбец value из spring_courses после слияния
        spring_courses = spring_courses.drop(columns="value")

        # Удаляем уже полностью зачисленных студентов
        df_spring = df_spring.merge(spring_courses.loc[(spring_courses["course1"] != "???")
            & (spring_courses["course2"] != "???")]["id"],
            how="left", on="id", indicator=True)
        df_spring = df_spring[df_spring["_merge"] == "left_only"].reset_index()\
            .drop(columns=["index", "_merge"])

# Сохраняем .csv файл
spring_courses.to_csv("res/res_spring.csv", index = False)
```
